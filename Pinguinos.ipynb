## Datos seleccionados: Penguins Data Set
Los datos de Palmer Penguins fueron recopilados y puestos a disposición por la Dra. Kristen Gorman en su trabajo en la Estación Palmer de la Antárctica LTER, miembre de la red de **Long Term Ecological Research Network**
El área de estudio de Palmer Long-Term Ecological Research (LTER) se encuentra al oeste de la Península Antártica y se extiende al sur y al norte de la cuenca Palmer desde tierra adentro hasta varios cientos de kilómetros mar adentro. Palmer Station es una de las tres estaciones de investigación de Estados Unidos ubicadas en la Antártida. Se encuentra en la isla Anvers, a mitad de camino de la Península Antártica, en la latitud 64,7 sur, longitud 64,0 oeste. Se puede ver una vista desde la estación en la cámara web de Palmer Station.

Datos iniciales: Los datos seleccionados corresponden a un censo de pinguinos donde se registraron los siguientes datos: **Especie**, **Localidad**, **Largo del culmen**, **Ancho del culmen**, **Largo de las aletas**, **Peso** y **Género**.

La dimensión de los datos es de 344 filas con 7 columnas, las cuales corresponden a los campos:Especie, Localidad, Ancho del pico, Largo del pico, Largo de las aletas, Peso y Género. Contamos con campos de tipo **object** y **float64**.
Los datos se encuentran balanceados, lo que indican que no hay clases con pocos datos, como tampoco existen datos nulos o blancos significativos, lo que se puede observar cuando se completa un archivo Excel con el dataset. Existen registros en blancos, pero solo correponden a 0,58% del total de la muestra.

#### e) Matrices de correlación

Acorde al gráfico que se presenta abajo, es posible determinar que, si bien la especie Gentoo tiene el culmen menos profundo, algunos de ellos poseen las aletas más largas de todos los pinguinos en esta relación de variables, y parece tener una correlación más marcada que las otras 2 especies.
En cuanto a la especie Chinstrap, parece  tener una correlación débil al relacionar estas 2 variables.

Pese a ello, en lo que es la profundidad del culmen para la especie Adelie, es un caso similar a los Chinstrap (ya que la elipse esta muy cercana a estar distribuida normalmente), aunque ligeramente mayor en cuanto a su correlación, con iteraciones de datos un poco más dispersas que las otras 2 especies.

```python
sns.jointplot(x="culmen_depth_mm", y="flipper_length_mm", hue="species", data=df, kind="scatter");
```

Al contrastar si existe correlación o no entre el largo del culmen y el largo de las aletas, se puede divisar que algo notoria en ellas es en la especie Gentoo, a su vez estando en un espectro mayor de las susodichas características, mostrandose que a mayor largo de culmen, mayor es el tamaño de sus aletas.

Respecto a la especie Chinstrap, parece mostrar una ligera correlación entre la profundidad del culmen y el largo de sus aletas, sumado a que muestra una mayor dispersión.
Finalmente, en la especie Adelie, solo se puede ver una correlación en lo que es el largo de las aletas, sin embargo el gráfico señala que no está vinculada al largo de su culmen.

```python
 sns.jointplot(x="culmen_length_mm", y="flipper_length_mm", hue="species", data=df, kind="scatter");
```

En el siguiente gráfico de correlación, es posible divisar que en lo que es el largo de aletas vs masa corporal de los especímenes, Adelie y Chinstrap están en unas medidas y correlación similares en lo que es un tamaño menor de la muestra en ambas características (pese a que los Adelie tienen una relación algo más débil, y por ende mas), mientras que la especie Gentoo esta en un peso y tamaño mayor de aletas, aunque su peso casi esté alineado con los ejes, muestra una relación algo más fuerte entre largo de aletas y masa corporal.

```python
sns.jointplot(x="body_mass_g", y="flipper_length_mm", hue="species", data=df, kind="scatter");
```

Visto de otra forma lo anterior, se puede analizar con una recta pendiente.

```python
g = sns.lmplot(x="flipper_length_mm",
               y="body_mass_g",
               hue="species",
               height=7,
               data=df,
               palette=['#2471A3','#FF5733','#FFC300'])
g.set_xlabels('Largo de aletas')
g.set_ylabels('Peso');
```

Según la siguiente vista, se puede determinar que, según su dispersión en el gráfico, para la especie Adelie y Gentoo existe una relación más fuerte y lineal respecto al largo del culmen, vs su masa corporal. Sin embargo, los Adelie, estan en un espectro de menor tamaño y menor peso, indicando que pueden ser los de menor tamaño de la muestra.

En el caso de los Chinstrap, se puede visualizar que hay una menor relación entre el largo del culmen y masa corporal debido a que los datos estan más dispersos.

```python
sns.jointplot(x="body_mass_g", y="culmen_length_mm", hue="species", data=df, kind="scatter");
```

En lo que relación largo de culmen y masa corporal respecta, se detecta que en la isla Dream hay una mayor variación de tamaños de culmen, que no indica claramente si se alimentan mejor o no en dicha isla, ya que estan en un rango de pesos menor y similar unos con otros.

En lo que a la isla Torgersen respecta, se puede divisar tambien unas muestras también algo dispersas (aunque menor que la isla Dream), pero ligeramente mas normalizada.

Finalmente, para la isla Biscoe, se detecta un mayor espectro de masa corporal vs largo de culmen, siguiendo una relación mas estrecha que las anteriormente comentadas y con mayores rangos de distribución.

```python
sns.jointplot(x="body_mass_g", y="culmen_length_mm", hue="island", data=df, kind="scatter");
```

En la siguente gráfica, se puede determinar que para la isla Torgersen las muestras están algo dispersas unas con otras, mientras que se quedan en un espectro pequeño, tanto para largo de aletas y peso.

Para la isla Dream, se puede determinar que su muestra de masas corporal vs largo de aletas esta mucho mas concentrada que en la Isla Torgersen, aunque en un espectro similar de tamaño y peso.

Finalmente, en lo que es la isla Biscoe, hay una concentración mayor de los datos en el espectro algo de las masas corporales/largo de aletas, sin embargo tienen mas valores atípicos dentro de los tamaños y pesos pequeños que las otras islas.

```python
sns.jointplot(x="body_mass_g", y="flipper_length_mm", hue="island", data=df, kind="scatter");
```

Como dato curioso, en el gráfico de correlación de peso/profundiad de culmen por isla, la isla Dream tiene datos dispersos en sus 2 extremos, en vez de tener una centralización de datos notoria como en el caso de Biscoe.

Finalmente, en una correlación entre el peso/largo de culmen por isla, se visualiza que en la isla Biscoe existe una tendencia y correlación de tener mayor peso y una mayor longitud de aletas en las especies, deduciendose que en esta isla se hallan los pinguinos de mayor embergadura, aunque teniendo varios datos atípicos.
